```python
def model(dataframe):
    clicks = []
    averages = []
    for index, item in dataframe.iterrows():
        sums = []
        click_holder = 0
        if item["keyboard"] == []:
            averages.append(np.inf)
        else:
            for thing in item["keyboard"]:
                sums.append(thing["t"])
            diffs = [x - sums[i - 1] for i, x in enumerate(sums)][1:]
            averages.append(np.mean(diffs))
        if item["mouse"] == []:
            clicks.append(0)
        else:
            for thing in item["mouse"]:
                if thing["event"] == "click":
                    click_holder += 1
            clicks.append(click_holder)
    dataframe["average"] = averages
    dataframe["clicks"] = clicks
    
    predictions = ""
    for index, item in dataframe.iterrows():
        if item["mouse"] == [] and item["keyboard"] == []:
            two = True
        else:
            two = False
            mouse_moves = []
            for thing in item["mouse"]:
                mouse_moves.append(thing["event"])
        if item["average"] <= 35 or "move" not in mouse_moves or two:
            #predictions.append(2)
            predictions += "2"
        elif item["clicks"] >= 2 or (item["keyboard"] and item["keyboard"][0]["key"] == 91):
            #predictions.append(1)
            predictions += "1"
        else:
            #predictions.append(0)
            predictions += "0"
    return predictions
            
            
```

```python
testJson = pd.read_json(path_or_buf="week3_test_events_final.jsonl", lines=True)
```

```python
preds = model(testJson)
```

```python
url = "https://class.disfraud.com/project3_score?predictions=" + preds
```
